## Descrição da Atividade:
Este notebook utiliza dados históricos das Olimpíadas (1896–2022) e dados
dos Jogos Olímpicos de Paris 2024 para consolidar o total de medalhas
por país, por meio da arquitetura por zonas.

Os dados foram obtidos das seguintes fontes:

- Base dos Dados – Histórico das Olimpíadas
- Kaggle – Paris 2024 Olympic Summer Games

## Objetivo DESSA Pipeline:
1. Limpeza + normalização de arquivos .csv em BRONZE
2. Geração de arquivos .parquet + metadados para a camada SILVER 
3. Fazer o JOIN das fontes e salvar como parquet na camada SILVER

## Importação das Bibliotecas:

Importações Default:

```python
import pandas as pd
import sys
from pathlib import Path
```

Configuração de Path:

```python
root = Path().absolute().parent.parent
sys.path.append(str(root))
```

Importações de módulos locais:

```python
from src import CatalogManager
from src.processing import clean_history_data, clean_paris_data, normalize_countries
from src.pipeline_utils import save_and_catalog
```

## 1. Definição das bases

```python
catalog = CatalogManager()
```

```python
bronze_base = root / "bronze"
silver_out = root / "silver"
```

```python
history_dir = bronze_base / "olympics_history_1896-2022"
paris_dir = bronze_base / "olympics_paris_2024"
```

### 2. Processando: Histórico 1896 - 2022

```python
csv_files = history_dir.glob("*.csv")

for file_path in csv_files:
  filename = file_path.stem
  df = pd.read_csv(file_path)

  df = normalize_countries(df)

  if "Medal_Tally" in filename:
        df = clean_history_data(df)
        desc = "Quadro de medalhas histórico com normalização de países e estações."
  else:
      desc = f"Dados históricos de {filename} normalizados."

  save_and_catalog(
        df=df,
        filename=filename,
        layer="silver",
        output_path=str(silver_out / "olympics_history"),
        description=desc,
        src=str(file_path),
        file_format=".parquet",
        catalog_manager=catalog,
        observations='Dado gerado na 2º etapa da pipeline - Limpeza e normalizaçaõ dos dados brutos em bronze para a camada silver'
    )
```

### 2.1 Processando: Paris 2024

```python
csv_files = paris_dir.glob("*.csv")

for file_path in csv_files:
  filename = file_path.stem
  df = pd.read_csv(file_path)

  df = normalize_countries(df)

  if "medals_total" in filename or "medallists" in filename:
        df = clean_paris_data(df)
        desc = f"Dados de Paris 2024 ({filename}) com colunas de medalhas padronizadas."
  else:
      desc = f"Dados de Paris 2024 ({filename}) normalizados."

  save_and_catalog(
        df=df,
        filename=filename,
        layer="silver",
        output_path=str(silver_out / "olympics_paris"),
        description=desc,
        src=str(file_path),
        file_format=".parquet",
        catalog_manager=catalog,
        observations='Dado gerado na 2º etapa da pipeline - Limpeza e normalizaçaõ dos dados brutos em bronze para a camada silver'
    )
```
